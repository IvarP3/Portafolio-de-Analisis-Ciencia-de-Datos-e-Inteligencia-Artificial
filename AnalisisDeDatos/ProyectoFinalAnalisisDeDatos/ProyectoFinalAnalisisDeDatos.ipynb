# LabelEncoder y OneHotEncoder

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
import pandas as pd

# 1. Lista de columnas para LabelEncoder (excluyendo 'Carrera')
label_cols = [
    'Genero',
    'Administraciondelcolegio',
    'AccesoDispositivos',
    'Celular',
    'InternetCelular',
    'Meet',
    'Periodo',
    'Observacion',
    'Ciudad',
    'PreferenciaClases',
    'PlataformaVirtual',
    'Zona',
    'TipoDispositivo'
]

le_mappings = {}

# 2. Aplicar LabelEncoder
for col in label_cols:
    le = LabelEncoder()
    df_enc[col] = le.fit_transform(df_enc[col].astype(str))
    le_mappings[col] = dict(zip(le.classes_, le.transform(le.classes_)))
    print(f"Mapeo de {col}: {le_mappings[col]}")

# 3. Aplicar OneHotEncoder a 'Carrera'
ohe = OneHotEncoder(sparse_output=False)
carrera_encoded = ohe.fit_transform(df_enc[['Carrera']])
carrera_df = pd.DataFrame(carrera_encoded, columns=ohe.get_feature_names_out(['Carrera']))
carrera_df.index = df_enc.index  # Asegurar que el 铆ndice coincide

# 4. Reemplazar 'Carrera' en el dataframe original
df_enc = pd.concat([df_enc.drop(columns=['Carrera']), carrera_df], axis=1)

# 5. Verificar resultado
df_enc.head()
```

Se utiliz贸 LabelEncoder en la mayor铆a de las columnas por ser binarias o con pocas categor铆as, permitiendo una codificaci贸n compacta sin aumentar la dimensionalidad. Aunque otras columnas nominales tambi茅n podr铆an codificarse con OneHotEncoder, se descartaron por tener muchas combinaciones o categor铆as poco estandarizadas que generar铆an demasiadas columnas. Se eligi贸 煤nicamente Carrera para OneHotEncoder por su importancia sem谩ntica, su car谩cter nominal sin orden y su n煤mero reducido de categor铆as, lo que mejora la representaci贸n sin comprometer la eficiencia del modelo.

# Clasificacion con normalizacion

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

# Divisi贸n del dataset para clasificaci贸n
X_train_c, X_test_c, y_train_c, y_test_c = train_test_split(
    X_normalizado, y_observacion, test_size=0.3, random_state=42, stratify=y_observacion)

# Mostrar clases presentes
print("Clases en y_train_c:", np.unique(y_train_c))
print("Clases en y_test_c:", np.unique(y_test_c))

# Modelos de clasificaci贸n
modelos_clf = {
    "Random Forest": RandomForestClassifier(random_state=42),
    "KNN Classifier": KNeighborsClassifier(),
    "Logistic Regression": LogisticRegression(multi_class='multinomial', max_iter=1000),
    "SVM": SVC(),
    "Decision Tree": DecisionTreeClassifier(random_state=42)
}

# Guardar m茅tricas
metricas_clf = {"Modelo": [], "Accuracy": [], "Precisi贸n": [], "Recall": [], "F1": []}
matrices_confusion = {}

# Entrenamiento y evaluaci贸n
for nombre, modelo in modelos_clf.items():
    modelo.fit(X_train_c, y_train_c)
    y_pred = modelo.predict(X_test_c)

    acc = accuracy_score(y_test_c, y_pred) * 100
    prec = precision_score(y_test_c, y_pred, average='macro') * 100
    rec = recall_score(y_test_c, y_pred, average='macro') * 100
    f1 = f1_score(y_test_c, y_pred, average='macro') * 100

    metricas_clf["Modelo"].append(nombre)
    metricas_clf["Accuracy"].append(acc)
    metricas_clf["Precisi贸n"].append(prec)
    metricas_clf["Recall"].append(rec)
    metricas_clf["F1"].append(f1)
    matrices_confusion[nombre] = confusion_matrix(y_test_c, y_pred)

    print(f" {nombre}")
    print(f"  Accuracy: {acc:.2f}%")
    print(f"  Precisi贸n: {prec:.2f}%")
    print(f"  Recall: {rec:.2f}%")
    print(f"  F1-Score: {f1:.2f}%\n")

# Mostrar m茅tricas
df_metricas_clf = pd.DataFrame(metricas_clf)
display(df_metricas_clf)

# Graficar matrices de confusi贸n
fig, axs = plt.subplots(1, 5, figsize=(24, 4))
for i, (nombre, cm) in enumerate(matrices_confusion.items()):
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', ax=axs[i])
    axs[i].set_title(f'Confusi贸n: {nombre}')
    axs[i].set_xlabel('Predicci贸n')
    axs[i].set_ylabel('Real')
plt.tight_layout()
plt.show()
```

Los resultados muestran que el modelo Random Forest ofrece el mejor desempe帽o general con una precisi贸n del 92.96% y excelente balance entre precisi贸n, recall y F1, lo que indica una alta capacidad de generalizaci贸n en esta tarea de clasificaci贸n.
